# Importing required libraries

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

# Loading training data from csv file and cleaning.

```python
titanic_ds=pd.read_csv('titanic_x_y_train.csv').copy()
titanic_ds.head(15)
```

```python
titanic_ds.isnull().sum()
```

```python
Pnames=[]
for  i in titanic_ds.Name.values:
    if i.split(',')[1].split('.')[0].strip()=='Mrs':
        Pnames.append(3)
    elif i.split(',')[1].split('.')[0].strip()=='Mr':
        Pnames.append(2)
    elif i.split(',')[1].split('.')[0].strip()=='Miss':
        Pnames.append(1)
    elif i.split(',')[1].split('.')[0].strip()=='Master':
        Pnames.append(0)
    else:
        Pnames.append(4)
Pnames=np.array(Pnames,dtype=int)
Pnames
```

```python
titanic_ds.describe(include='all')
```

```python
titanic_ds.loc[titanic_ds['Sex']=='female','Sex']=0
titanic_ds.loc[titanic_ds['Sex']=='male','Sex']=1
```

```python
titanic_ds.loc[titanic_ds['Embarked'].isnull(),'Embarked']='S'
titanic_ds.loc[titanic_ds['Embarked']=='S','Embarked']=1
titanic_ds.loc[titanic_ds['Embarked']=='C','Embarked']=2
titanic_ds.loc[titanic_ds['Embarked']=='Q','Embarked']=3
titanic_ds.loc[titanic_ds['Cabin'].isnull(),'Cabin']=0
titanic_ds.loc[titanic_ds['Cabin']!=0,'Cabin']=1

```

```python
titanic_ds.loc[titanic_ds['Age'].isnull(),'Age']=titanic_ds['Age'].median(skipna=True)
```

```python
titanic_ds.drop('Name',axis=1,inplace=True)
titanic_ds.drop('Ticket',axis=1,inplace=True)
```

```python
titanic_ds.head(10)
```

```python
X_train=titanic_ds.values[:,:-1]
X_train=np.append(X_train,Pnames.reshape(-1,1),axis=1)
X_train.shape
```

```python
Y_train=titanic_ds.values[:,-1]
Y_train
```

```python
print(X_train.shape,Y_train.shape)
```

# Loading test data and cleaning it.

```python
test_ds=pd.read_csv('titanic_x_test.csv').copy()
test_ds.head()
```

```python
Test_Pnmaes=[]
for i in test_ds.Name.values:
    if i.split(',')[1].split('.')[0].strip()=='Mrs':
        Test_Pnmaes.append(3)
    elif i.split(',')[1].split('.')[0].strip()=='Mr':
        Test_Pnmaes.append(2)
    elif i.split(',')[1].split('.')[0].strip()=='Miss':
        Test_Pnmaes.append(1)
    elif i.split(',')[1].split('.')[0].strip()=='Master':
        Test_Pnmaes.append(0)
    else:
        Test_Pnmaes.append(4)
Test_Pnmaes=np.array(Test_Pnmaes,dtype=int)
Test_Pnmaes
    
```

```python
test_ds.loc[test_ds['Sex']=='female','Sex']=0
test_ds.loc[test_ds['Sex']=='male','Sex']=1
test_ds.loc[test_ds['Embarked'].isnull(),'Embarked']='S'
test_ds.loc[test_ds['Embarked']=='S','Embarked']=1
test_ds.loc[test_ds['Embarked']=='C','Embarked']=2
test_ds.loc[test_ds['Embarked']=='Q','Embarked']=3
test_ds.loc[test_ds['Cabin'].isnull(),'Cabin']=0
test_ds.loc[test_ds['Cabin']!=0,'Cabin']=1
```

```python
test_ds.loc[test_ds['Age'].isnull(),'Age']=test_ds['Age'].median(skipna=True)
test_ds.drop('Name',axis=1,inplace=True)
test_ds.drop('Ticket',axis=1,inplace=True)
```

```python
X_test=test_ds.values
X_test=np.append(X_test,Test_Pnmaes.reshape(-1,1),axis=1)
X_test.shape
```

```python
X_test.shape
```

# Logistic Regression

```python
clf=LogisticRegression(solver='saga',max_iter=100000, tol=0.00001)
clf.fit(X_train,Y_train)
```

```python
clf.score(X_train,Y_train)
```

```python
predicted=clf.predict(X_test)
```

```python
np.savetxt(X=predicted, delimiter=',', fname='titanic_ds predictions.csv', fmt='%.0f')
```
